```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
# Load the data of the book
with open('TheDaVinciCode.txt', 'r') as f:
    data = f.read()
    data = data.replace("\n", " ")
    data = data.replace("\t", " ")
    data = data.replace("\r", " ")
    data = data.replace("  ", " ")
    data = data.replace("   ", " ")
    data = data.replace("    ", " ")
    data = data.replace("     ", " ")
```

```python
# Get every chapter in the data with case sensitivity.
paragraphs = data.split("CHAPTER")
print(len(paragraphs))

# Merge the paragraphs if the first element is not a number.
tmp = []
for i in range(1, len(paragraphs)-1):
    if not paragraphs[i][1].isdigit():
        tmp.append(paragraphs[i-1] + paragraphs[i])
        paragraphs.remove(paragraphs[i])
    
print(i)

print(paragraphs[106])
```

```python
persons = ['robert', 'grand master', 'sophie', 'langdon', 'mr. langdon', 'jacques sauniere', 'sauniere', 'andre', 'marie chauvel', 'jacques', 'master masons', 'roslin', 'leigh teabing', 'robert langdon']
i=1
for text in paragraphs:
    print('------------------ CHAPTER ' + str(i) + ' ------------------')
    # Tokenize the text into words
    words = word_tokenize(text)

    # Tag the words with their corresponding parts of speech
    tagged_words = nltk.pos_tag(words)

    # Perform NER on the tagged words
    ne_chunked_words = nltk.ne_chunk(tagged_words)

    persons = []
    previous = ''
    # Print the named entities
    for chunk in ne_chunked_words:
        if hasattr(chunk, 'label'):
            current = ' '.join(c[0] for c in chunk)
            if current != previous:
                if chunk.label() != 'PERSON':
                    print(chunk.label(), ' '.join(c[0] for c in chunk))
                    previous = current
                else:
                    if current.lower() in persons:
                        print(chunk.label(), ' '.join(c[0] for c in chunk))
                        previous = current
            if chunk.label() == 'PERSON':
                persons.append(' '.join(c[0] for c in chunk))
    i += 1
```

```python
persons = list(set(persons))    
print(persons)
```
