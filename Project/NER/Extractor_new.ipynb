```python
import re
import pickle
```

```python
with open('C:/Users/Admin/Documents/University/CurrentClass/IRTM/DaVinciCodeTheTrackOfRobertLangdon/Project/TheDaVinciCode.txt', 'r') as f:
    data = f.read()
    data = data.replace("\n", " ")
    data = data.replace("\t", " ")
    data = data.replace("\r", " ")
    data = data.replace("  ", " ")
    data = data.replace("   ", " ")
    data = data.replace("    ", " ")
    data = data.replace("     ", " ")
    ' '.join(data.split())

f.close()
```

```python
# data to list of sentences
sentences = re.split("(?<!\w\.\w.)(?<![A-Z]\.)(?<![A-Z][a-z]\.)(?<=\.|\?)|!", data) 
print('number of sentences: ', len(sentences))
```

```python
sentences = sentences[:300]
```

## Roberta

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
from tqdm import tqdm

print('import done')

# Load the pre-trained model and the tokenizer
robertaTokenizer = AutoTokenizer.from_pretrained("Jean-Baptiste/roberta-large-ner-english")
print('tokenizer loaded')
roberta_ner = AutoModelForTokenClassification.from_pretrained(
            "Jean-Baptiste/roberta-large-ner-english")

print('loader done')

# Create a NER pipeline consisting of the model and it's tokenizer
roberta_ner_model = pipeline('ner', model=roberta_ner, tokenizer=robertaTokenizer, aggregation_strategy="simple")

print('pipeline done')
```

```python
roberta_tags = []

# Parse sentences one by one
counter = 1
for sentence in tqdm(sentences):
    result = roberta_ner_model(sentence)
    for entity in result:
      # Collect words and tags
      roberta_tags.append((entity['word'], entity['entity_group'], counter))
    counter+=1

# Print the word-NER tag tuples for first episode
roberta_tags
```

## Bert

```python
# Load the pre-trained model and the tokenizer
bertTokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
bert_ner = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

print('loader done')

# Create a NER pipeline consisting of the model and it's tokenizer
bert_ner_model = pipeline('ner', model=bert_ner, tokenizer=bertTokenizer, aggregation_strategy="simple")

print('pipeline done')
```

```python
bert_tags = []

# Parse sentences one by one
counter = 1
for sentence in tqdm(sentences):
    result = bert_ner_model(sentence)
    for entity in result:
      # Collect words and tags
      bert_tags.append((entity['word'], entity['entity_group'], counter))
    counter+=1

# Print the word-NER tag tuples for first episode
bert_tags
```
