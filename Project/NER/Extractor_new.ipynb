```python
import re
import pickle
```

```python
with open('TheDaVinciCode.txt', 'r') as f:
    data = f.read()
    data = data.replace("\n", " ")
    data = data.replace("\t", " ")
    data = data.replace("\r", " ")
    data = data.replace("  ", " ")
    data = data.replace("   ", " ")
    data = data.replace("    ", " ")
    data = data.replace("     ", " ")
    ' '.join(data.split())

f.close()
```

```python
# data to list of sentences
sentences = re.split("(?<!\w\.\w.)(?<![A-Z]\.)(?<![A-Z][a-z]\.)(?<=\.|\?)|!", data) 
print('number of sentences: ', len(sentences))
```

```python
sentences = sentences[:300]
```

## Roberta

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
from tqdm import tqdm

print('import done')

# Load the pre-trained model and the tokenizer
tokenizer = AutoTokenizer.from_pretrained(
            "Jean-Baptiste/roberta-large-ner-english")
print('tokenizer loaded')
bert_ner = AutoModelForTokenClassification.from_pretrained(
            "Jean-Baptiste/roberta-large-ner-english")

print('loader done')

# Create a NER pipeline consisting of the model and it's tokenizer
ner_model = pipeline('ner', model=bert_ner, tokenizer=tokenizer, aggregation_strategy="simple")

print('pipeline done')
```

```python
bert_tags = []

# Parse sentences one by one
counter = 1
for sentence in tqdm(sentences):
    result = ner_model(sentence)
    for entity in result:
      # Collect words and tags
      bert_tags.append((entity['word'], entity['entity_group'], counter))
    counter+=1

# Print the word-NER tag tuples for first episode
bert_tags
```
