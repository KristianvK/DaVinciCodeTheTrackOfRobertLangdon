```python
f = open("TheDaVinciCode.txt", "r")

data = f.read()

f.close()
```

```python
import spacy
from spacy import displacy
from collections import Counter
# import en_core_web_sm
# nlp = en_core_web_sm.load()
nlp = spacy.load('en_core_web_sm')
```

```python
data = data.replace("\n", " ")
data = data.replace("\t", " ")
data = data.replace("\r", " ")
data = data.replace("  ", " ")
data = data.replace("   ", " ")
data = data.replace("    ", " ")
data = data.replace("     ", " ")
data_beginning = data[:100000]
data_beginning
```

```python
doc = nlp(data_beginning)
print([(X.text, X.label_) for X in doc.ents])
```

```python
person = []
for ent in doc.ents:
    if ent.label_ == "PERSON" and ent.text not in person:
        print(ent.text)
        person.append(ent.text)
```

```python
displacy.render(nlp(data_beginning), jupyter=True, style='ent')
```
